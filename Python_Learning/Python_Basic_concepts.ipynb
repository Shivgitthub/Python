## Exception_Handling

1. ValueError - The Value of that variable is incorrect to the variable's nature so it rises an error 

example : if variable stores only int values if we use any non integar values like string,float,bool,complex_num it throws valueError


2. NameError - This error arises In a function when that variable is call to function that variable is not assigned to any value

2. Syntaxerror- This error arires when function use case syntax is mismatches with original function syntax

```python
def main():
    x = get_int(" what's x is ")
    print(f'x is {x}')

def get_int(prompt):

    while True:
        try:
            return  int(input(prompt))
        except ValueError:
            pass

print(f' x is {x}')
```
