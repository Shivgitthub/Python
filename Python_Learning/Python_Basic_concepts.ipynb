```python
def num_outliers(data:pd.DataFrame,criteria='gaussian',n=3):

    if criteria == 'gaussian':

        ## Number of outliers given data below = data < (data.mean - n*data.std()) and above = data > (data.mean()+n*data.std())

        low = len(data[data < (data.mean() - n*data.std())])
        high = len(data[data > (data.mean() + n*data.std())])
        total = low+high
        return total,low,high
    elif criteria == 'whisker':
        M_f = 1.5
        Q1 = 0.25
        Q3 = 0.75
        # Removing the outliers by IQR Method which quaritles are play role to identify the how many outliers are present
        """
        IQR = Q3-Q1
        LB= Q1 - 1.5 IQR
        UB = Q3 + 1.5 IQR
        In here 1.5 is M_f
        low = data < LB
        high = data > UB
        low+high --> Total Outliers in data
        """
        

        low = len(data[data < data.quantile(Q1) - (M_f *(data.quantile(Q3) - data.quantile(Q1) ))])
        high = len(data[data > data.quantile(Q3) + (M_f *(data.quantile(Q3) - data.quantile(Q1) ))])
        total = low+high
        return total,low,high
```

```python
def factorial(x):
    """
    Description: 
    Function for finding factorial for given number

    Parameter:
    1. x:int
    The number is to find factorial value

    Result: int
    Function returns factorial value for given number   
    """

    if x < 0:
        return None
    elif x == 0 or x == 1:
        return 1
    else:
        return x * factorial(x-1)  
```

```python
def extending_list(list1:list,list2:list)->list:

    """
    Description:
    Takes list-1 and extends it with elements of list-2
    Such that len(result_list) = len(list-1)+len(list-2)

    parameter:
    list1:list - Original list to be extended
    list2:list - list of elements to be added to original list

    Result:list
    Result of extension of given two list
    """

    assert type(list1) == list and type(list2) == list, "Only List must be allowed"

    new_list = list1[:]

    new_list.extend(list2)

    assert len(new_list) == len(list1)+len(list2), "Output result is not expected according to logic"

    return new_list
```

```python
def strange_operation(iterable:pd.Series):

    """
    Description : This funciton helps to find ration of standard deviation and variation of given series and multiplied by every iteration of given series

    parameter:
    iterable : pd.Series

    Result: pd.Series
    It returns multliplied ration to every member of given series
    """

    variance = iterable.var()
    standard_deviation = iterable.std()

    ratio = variance/standard_deviation

    return iterable.map(lambda x: x*ratio)
```

### 2. lru_cache()

Memoizing callable and returns the stored value if the function is called with the same arguments again. It can save time when an expensive or I/O bound function is periodically called with the same arguments

```python

# function without lru_cache

def calc_uncached_factorial(n):
    if n<=1:
        return 1
    return n * calc_uncached_factorial(n-1)
```

```python
%timeit calc_uncached_factorial(10)
```

```python

# function with lru_cache
@functools.lru_cache(maxsize=None)
def calc_cached_factorial(n):
    if n<=1:
        return 1
    return n * calc_cached_factorial(n-1)
```

```python
%timeit calc_cached_factorial(10)
```

```python
def mean(iterable):
    return sum(iterable)/len(iterable)

def variance(iterable):
    avg = mean(iterable)
    diff_sq = list(map(lambda x: (x-avg)**2 ,iterable))
    return sum(diff_sq)/len(iterable)

def standard_deviation(iterable):
    return variance(iterable) ** (1/2)

def descriptives(function_list,iterable):
    """
    Runs all the funcitons in function list over the iterable and returns result list in same order
    """
    return list(map(lambda x: x(iterable),function_list))
```

```python
def adder(x,y):
    return x+y

def subractor(x,y):
    return x-y

def multiplicator(x,y):
    return x*y

def dividor(x,y):
    return x/y

def square(a):
    return a**(1/2)
```

```python
import functools
```

```python
def herons_adder(*values):
    """ Can add any number of values (herons formula required 3 times to add)"""
    return functools.reduce(adder,values)
```

```python
def herons_multipier(*values):
    """ Can Multiply any number of values (herons formula required 4 times to multple)"""
    return functools.reduce(multiplicator,values)
```

### Function to perform Herons's Formula

```python
def herons_formula(a,b,c):


    semi_parameter = dividor(herons_adder(a,b,c),2)

    a_difference = subractor(semi_parameter,a)
    b_difference = subractor(semi_parameter,b)
    c_difference = subractor(semi_parameter,c)

    return square(herons_multipier(semi_parameter,a_difference,b_difference,c_difference))
```

```python
herons_formula(3,4,6)
```
