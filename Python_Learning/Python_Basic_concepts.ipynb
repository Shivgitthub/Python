#### `File_Handling`

- There are Some modes that files can read and write data that are use case of the file handling

1.  `r` --> Read Mode
2. `w` --> Write Mode
3. `a` --> append Mode

```python
file = open('sample.txt','r')

f = file.readlines()

print(f)
```

```python
## Removiing the \n and unwanted white spaces and \ special characters from file there fore using loops that can avoid

new_list = []
for line in f:
    if line[-1] == '\n':
        new_list.append(line[:-1])
    else:
        new_list.append(line)

print(new_list)
```

```python
new_list = []
for line in f:
    new_list.append(line.strip())

print(new_list)
```

```python
import pandas as pd
```

```python
def num_outliers(data:pd.DataFrame,criteria='gaussian',n=3):

    if criteria == 'gaussian':

        ## Number of outliers given data below = data < (data.mean - n*data.std()) and above = data > (data.mean()+n*data.std())

        low = len(data[data < (data.mean() - n*data.std())])
        high = len(data[data > (data.mean() + n*data.std())])
        total = low+high
        return total,low,high
    elif criteria == 'whisker':
        M_f = 1.5
        Q1 = 0.25
        Q3 = 0.75
        # Removing the outliers by IQR Method which quaritles are play role to identify the how many outliers are present
        """
        IQR = Q3-Q1
        LB= Q1 - 1.5 IQR
        UB = Q3 + 1.5 IQR
        In here 1.5 is M_f
        low = data < LB
        high = data > UB
        low+high --> Total Outliers in data
        """
        

        low = len(data[data < data.quantile(Q1) - (M_f *(data.quantile(Q3) - data.quantile(Q1) ))])
        high = len(data[data > data.quantile(Q3) + (M_f *(data.quantile(Q3) - data.quantile(Q1) ))])
        total = low+high
        return total,low,high
```

```python
def gaussian_outliers(data,n):
    ## Number of outliers given data below = data < (data.mean - n*data.std()) and above = data > (data.mean()+n*data.std())

    low = len(data[data < (data.mean() - n*data.std())])
    high = len(data[data > (data.mean() + n*data.std())])
    total = low+high
    return total,low,high

def whisker_outliers(data,M_f):
    # M_f = 1.5
    Q1 = 0.25
    Q3 = 0.75
    # Removing the outliers by IQR Method which quaritles are play role to identify the how many outliers are present
    """
    IQR = Q3-Q1
    LB= Q1 - 1.5 IQR
    UB = Q3 + 1.5 IQR
    In here 1.5 is M_f
    low = data < LB
    high = data > UB
    low+high --> Total Outliers in data
    """
    IQR = data.quantile(Q3) - data.quantile(Q1)
    low = len(data[data < data.quantile(Q1) - (M_f *(IQR))])
    high = len(data[data > data.quantile(Q3) + (M_f *(IQR))])
    total = low+high
    return total,low,high

def find_outliers(data:pd.DataFrame,gaussian=True ,n=3,M_f=1.5):
    return gaussian_outliers(data,n) if gaussian else whisker_outliers(data,M_f)
    
```

### Bugs

- Types of Bugs
    1. Overt 
    2. Covert 

A. Overt Bugs --> are visible/ easily to identify , obvious to catch, code crashes or runs forever

B. Covert Bugs --> are not obvious to catch , because it does not crashes the code but it returns abnormal value / absurd, not relative value what we expected
    1. implementation of logic for function is flawed 
    2. Some variable mishandled 
    
**Consistency of Bugs**

    1. Persistent (High_reproducible)--> Persistent  bugs appear everytime the code is in run if return value of funciton is wrong so that we can identified that funciton variable / somes values wrongly misplaced
    2. Intermittent (Low_reproducible) --> Intermittent bugs are only visible under special circumstances


Categories of Bugs

1. overt and Persistent 
    - Obvious to catch/detect, Reproducible
    - Defensive programming ensuers bug fall into this category

2. Overt and intermittent
    - Harder to debug , Somewhat reproducible
    - if spotted, can be handled

3. Covert and Intermittent
    - Highly harder to find the bug
    - Takes a long time to even detect bug
