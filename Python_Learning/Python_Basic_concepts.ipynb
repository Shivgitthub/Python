#### `File_Handling`

- There are Some modes that files can read and write data that are use case of the file handling

1.  `r` --> Read Mode
2. `w` --> Write Mode
3. `a` --> append Mode

```python
file = open('sample.txt','r')

f = file.readlines()

print(f)
```

```python
## Removiing the \n and unwanted white spaces and \ special characters from file there fore using loops that can avoid

new_list = []
for line in f:
    if line[-1] == '\n':
        new_list.append(line[:-1])
    else:
        new_list.append(line)

print(new_list)
```

```python
new_list = []
for line in f:
    new_list.append(line.strip())

print(new_list)
```

```python
file.close()  ### Closing file is important because it can't save the changes in that file
```

```python
### Write Mode with open keyword
file = open('file.txt','w')

file.write('Python Learning is fun \n \n')

file.write('Now in python file handling was learning right now')

file.close()
```

```python
# w is write mode, In write mode if write more than once then file replaced with new data and old data is erased

# a is append mode which means new data is updated on the file without distrubin' the old data

name = input('What is your name ? ')

file = open('dummy_repeatation.txt','a')

file.write(f'{name} \n')

file.close()
```

### **`with`** key words

Using with keyword we can able to open and close the file with out using close method for that file

```python
name = input('What is your name? ')
with open('dummy_repeatation.txt','a') as file :
    file.write(f'{name} \n')
```

```python
## Reading the existing the file 

with open('dummy_repeatation.txt','r') as file:
    lines = file.readlines()

for line in lines:
    print('hello',line.rstrip())  
## If we don't use rstrip it gives extra two new lines becuase we delibrately write the data with \n 
```

```python
with open('dummy_repeatation.txt','r') as file:
    for line in sorted(file):
        print('hello',line,end='')
```

```python
names = []

with open('dummy_repeatation.txt','r') as fl:
    for line in fl:
        names.append(line.rstrip())

for name in names:
    print(f'hello, {name}')
```

```python
student = []



with open('dummy_repeatation.csv','r') as files:
    for line in files:
        name,house = line.rstrip().split(',')
        students = {'name':name,'house':house}
        student.append(students)

def get_name(students):
    return students['name']

def get_house(students):
    return students['house']

for i in sorted(student,key=get_house):
    print(f"{i['name']} is in {i['house']}")
```

```python
import csv

students = []

with open('dummy_repeatation.csv') as file:
    reader = csv.reader(file)
    for read in reader:
     #   print(read)
        student = {'name':read[0],'house':read[1]}
        students.append(student)

for row in sorted(students,key=lambda student: student['house']):
    print(f"{row['name']} is in {row['house']}")
```

## if use dict reader in csv we can write and read csv file in format of json it means csv first line must be heading / title or dict keys 

```python
s1 = []
with open('dummy_repeatation.csv') as file:
    reader = csv.DictReader(file)
    
    for row in reader:
        print(row)
        s2 = {'name':row['name'],'address':row['address']}
        s1.append(s2)

for row in sorted(s1,key= lambda s2:s2['name']):
    print(f" {row['name']} is in {row['address']}")
```
